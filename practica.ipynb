## SQLite con python

coneccion a una bd de sqlite3

```python
import sqlite3
conn = sqlite3.connect("prueba.db")

cursor = conn.cursor()
cursor.execute("CREATE TABLE IF NOT EXISTS usuarios (id INTEGER, nombre TEXT)")
cursor.execute("INSERT INTO usuarios VALUES (1, 'Juan')")
cursor.execute("SELECT * FROM usuarios")
result = cursor.fetchall()
print(result)
```

```python
import sqlite3

try:
    conn = sqlite3.connect('mi_base.db')
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS usuarios (id INTEGER, nombre TEXT)")
    cursor.execute("INSERT INTO usuarios VALUES (1, 'dato')")
    cursor.execute("SELECT * FROM usuarios")
    # ¡IMPORTANTE! Hacer commit
    conn.commit()
    result = cursor.fetchall()
    print(result)
    
except sqlite3.Error as error:
    print("Error:", error)
    
finally:
    # Siempre cerrar
    if cursor:
        cursor.close()
    if conn:
        conn.close()

```
